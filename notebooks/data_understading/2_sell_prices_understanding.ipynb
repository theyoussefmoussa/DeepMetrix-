```python
calendar = pd.read_csv(f'{BASE_PATH}/data/raw/calendar.csv')
sales = pd.read_csv(f'{BASE_PATH}/data/raw/sales_train_evaluation.csv')
```

```python
# Does every wm_yr_wk in price exist in calendar?
missing_weeks = set(price['wm_yr_wk']) - set(calendar['wm_yr_wk'])
print(f"wm_yr_wk in price but not calendar: {len(missing_weeks)}")

# Does every (item_id, store_id) in sales have at least one price entry?
sales_keys = set(zip(sales['item_id'], sales['store_id']))
price_keys = set(zip(price['item_id'], price['store_id']))
missing_combos = sales_keys - price_keys
print(f"item-store combos in sales with zero price history: {len(missing_combos)}")
```
