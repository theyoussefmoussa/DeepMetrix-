```python
import numpy as np
import pandas as pd
import os
from dotenv import load_dotenv
load_dotenv()
BASE_PATH = os.environ.get("BASE_PATH")
```

```python
sales = pd.read_csv(f"{BASE_PATH}/data/raw/sales_train_evaluation.csv")
sales.head()
```

```python
print(f"Before Cleaning: {sales.memory_usage(deep=True).sum() / 1e6:.1f} MB")
```

```python
sales.info(verbose=True)
```

```python
# strip prefix and downcast d
sales.columns = sales.columns.str.replace(r'^d_', '', regex=True)

cols = sales.columns[sales.columns.str.isnumeric()]
sales[cols] = sales[cols].astype('Int16')
sales['item_id'] = sales['item_id'].astype('category')
sales.info()
```

```python
sales.rename(columns={'id' : 'item_store_id'}, inplace=True)
```

```python
sales['item_store_id'].nunique()
```

```python
sales['dept_id'] = sales['dept_id'].astype('category')
sales.info(verbose=True)
```

```python
sales['cat_id'] = sales['cat_id'].astype('category')
```

```python
sales.info(verbose=True)
```

```python
sales['store_id'] = sales['store_id'].astype('category')
```
