```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
BASE_PATH = os.environ.get("BASE_PATH")
calendar = pd.read_csv(f'{BASE_PATH}/data/raw/calendar.csv')
calendar.head()
```

```python
calendar.info()
```

```python
columns_to_rename = {
    'd' : 'day_number',
    'event_name_1' : 'event_name',
    'event_type_1' : 'event_type',
    'wm_yr_wk' : 'walmart_year_week'
}

calendar.rename(columns=columns_to_rename, inplace=True)
calendar.head()
```

```python
calendar['date'] = pd.to_datetime(calendar['date'])
```

```python
calendar['weekday'].head()
```

```python
calendar.drop(columns=['wday'], inplace=True)
calendar.info()
```

```python
calendar['walmart_year_week'].nunique()
```

```python
print(calendar['walmart_year_week'].max())
print(calendar['walmart_year_week'].min())
```

```python
calendar['walmart_year_week'] = calendar['walmart_year_week'].astype('int32')
```

```python
calendar.drop(columns=['event_name_2', 'event_type_2'], inplace=True)
```

```python
calendar.info()
```

```python
months = {
    1: "January",
    2: "February",
    3: "March",
    4: "April",
    5: "May",
    6: "June",
    7: "July",
    8: "August",
    9: "September",
    10: "October",
    11: "November",
    12: "December"
}

calendar['month_name'] = calendar['month'].map(months)
calendar.drop(columns=['month'], inplace=True)
```

```python
calendar.info()
```

```python
calendar['day_number'].head()
```

```python
calendar['day_number'] = calendar['day_number'].str.replace('d_', '')
calendar['day_number'] = calendar['day_number'].astype('int16')
```

```python
calendar.info()
```

```python
cols = ['event_type', 'event_name']
def fill_with_key(df, key, cols):
    for col in cols:
        df[col] = df[col].fillna(key)
    return df

calendar = fill_with_key(calendar,'No Event', cols)
calendar.isna().sum()
```

```python
calendar.info()
```

```python
calendar['event_name'].nunique()
```

```python
calendar['event_name'].unique()
```

```python
for col in cols: 
    calendar[col] = calendar[col].astype('category')
```

```python
calendar.info()
```

```python
binary_cols = ['snap_CA', 'snap_TX', 'snap_WI']

for col in binary_cols:
    calendar[col] = calendar[col].astype(bool)
calendar.info()
```

```python
calendar['event_type'].unique()
```

```python
calendar['weekday'] = calendar['weekday'].astype('category')
calendar.info()
```

```python
calendar['month_name'] = calendar['month_name'].astype('category')
calendar.info()
```

```python
calendar['year'].value_counts()
```

```python
calendar["date_year"] = calendar["date"].dt.year
calendar['date_year'].value_counts()
```

```python
calendar.drop(columns=['date_year', 'year'], inplace=True)
calendar.info()
```

```python
calendar.head()
```

```python
output_path = os.path.join(BASE_PATH, "data", "processed", "calendar_cleaned.parquet") # type: ignore

calendar.to_parquet(output_path, engine="pyarrow", index=False)
```
